# **Seaborn**

**Why Seaborn?**

- **Simpler Syntax**: Seaborn simplifies complex plots with easier syntax compared to Matplotlib.
- **Built-in Statistical Plots**: Seaborn includes functions for statistical plots like regression lines and distributions.
- **Aesthetics**: Seaborn has better default aesthetics, saving time on customization.
- **Pandas Integration**: Direct support for Pandas DataFrames, making it easy to plot data without restructuring.
- **Complex Plots**: Makes creating complex plots like heatmaps, pair plots, and facet grids easy.
- **Exploratory Data Analysis (EDA)**: Ideal for quickly exploring data relationships with minimal code.

Seaborn is useful for quick, high-quality visualizations, while Matplotlib is more customizable but requires more code.

https://seaborn.pydata.org/api.html

### **Boxplot**

A boxplot is a standardized way of displaying the distribution of data based on a five number summary (â€œminimumâ€, first quartile [Q1], median, third quartile [Q3] and â€œmaximumâ€). It can tell you about your outliers and what their values are. Boxplots can also tell you if your data is symmetrical, how tightly your data is grouped and if and how your data is skewed.

image.png

```python
sns.boxplot(tips,x="total_bill")
```

```python
sns.boxplot(tips,y="total_bill")
```

```python
# Box plot
sns.boxplot(data=tips,x='day',y='total_bill')
```

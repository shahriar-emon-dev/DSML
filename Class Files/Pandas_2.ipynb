---
###  **Use `pd.to_numeric()` when:**
- You're converting **strings** that **look like numbers** into actual numeric types.
- You want to **safely handle errors** like text in a numeric column.

---

###  Example:

```python
import pandas as pd

df = pd.DataFrame({"age": ["25", "30", "unknown"]})

# Force conversion with errors='coerce'
df["age"] = pd.to_numeric(df["age"], errors='coerce')
print(df)
```

#### Output:
```
     age
0   25.0
1   30.0
2    NaN
```

---

###  Why not just `.astype(int)`?

```python
df["age"] = df["age"].astype(int)  #  This will crash on "unknown"
```

- **`.astype()` is strict**  it fails if there are bad values.
- **`pd.to_numeric()` is flexible**  it can **coerce** bad values to `NaN`.

---

###  When to Use Which:

| Task                          | Use This              |
|-------------------------------|------------------------|
| Clean numeric strings         | `pd.to_numeric()`      |
| Safe conversion with errors   | `pd.to_numeric(errors='coerce')` |
| Youre 100% sure it's clean   | `.astype(int)` or `.astype(float)` |
| Converting to datetime        | `pd.to_datetime()`     |
| Converting multiple types     | `.astype({col: type})` |

---


---

##  **Pandas Data Type Conversion Cheat Sheet**

|  **Goal**                        |  **Use this**                                  |  **Extra Notes**                         |
|----------------------------------|--------------------------------------------------|--------------------------------------------|
| Convert to integer               | `df["col"] = df["col"].astype(int)`             | Only works if all values are clean ints    |
| Convert to float                 | `df["col"] = df["col"].astype(float)`           | Use for decimal numbers                    |
| Convert to string                | `df["col"] = df["col"].astype(str)`             | Works even with numbers                    |
| Convert to boolean              | `df["col"] = df["col"].astype(bool)`            | True = non-zero/non-empty                  |
| Convert multiple columns         | `df = df.astype({"col1": int, "col2": float})`  | Best way for bulk type fixes               |
| Convert to datetime              | `df["col"] = pd.to_datetime(df["col"])`         | Smart & safe for any date format           |
| Convert to datetime (with format)| `pd.to_datetime(df["col"], format="%d-%m-%Y")`  | Speeds up parsing                          |
| Convert to numeric safely        | `pd.to_numeric(df["col"], errors='coerce')`     | Converts bad values to NaN                 |
| Convert object to category       | `df["col"] = df["col"].astype("category")`      | Useful to save memory                      |

---

###  `.astype()` vs `pd.to_numeric()` vs `pd.to_datetime()`

| Function           | Safe for dirty data | Handles errors | Flexible formats |
|-------------------|---------------------|----------------|------------------|
| `.astype()`        |  No                |  No           |  No             |
| `pd.to_numeric()`  |  Yes               |  Yes          |  (only numbers) |
| `pd.to_datetime()` |  Yes               |  Yes          |  Yes            |

### **Merging/Joining**

````python
temp_df = pd.DataFrame({
'City': ['New York', 'Chicago', 'Orlando'],
'Temperature': [32, 30, 35]
})

humidity_df = pd.DataFrame({
'City': ['Chicago', 'New York', 'San Diego'],
'Humidity': [60, 65, 70]
})
````

````python
temp_df
````

````python
humidity_df
````

````python
df = pd.merge(temp_df, humidity_df, on = "City") ## Inner Join By Default
df
````

````python
df = pd.merge(temp_df, humidity_df, on = "City", how= "outer")
df
````

````python
df = pd.merge(temp_df, humidity_df, on = "City", how= "left")
df
````

````python
df = pd.merge(temp_df, humidity_df, on = "City", how= "right")
df
````

### **Apply a Function to a Single Column (Series)**

````python
def rating(x, high, low):
    if x > high:
        return "high"
    elif x > low:
        return "med"
    else:
        return "low"

sales_data["sales_amount"].apply(rating, args=(1800, 1000))
````

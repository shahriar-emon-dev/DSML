# Scatter Plots

image.png

- Bivariate Analysis
- numerical vs numerical
- Use case - Finding correlation

```python
tips= sns.load_dataset("tips")

tips.head()
```

```python
plt.scatter(tips['total_bill'],tips['tip'])
```

```python
plt.scatter(tips['total_bill'],tips['tip'],s=tips['size']*20)
```

```python
## size

plt.scatter(tips['total_bill'],tips['tip'],s=tips['tip']*10)
```

```python
# Map "sex" to a color, here we use a colormap
plt.scatter(
    tips['total_bill'],
    tips['tip'],
    s=tips['tip'] * 10,  # Size of the points based on tip value
    c=tips['sex'].map({'Male': 'blue', 'Female': 'red'}),  # Mapping 'sex' to color
    alpha=0.7  
)

plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Scatter plot of Tip vs Total Bill')
plt.show()
```

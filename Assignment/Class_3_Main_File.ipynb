**Why Do We Need Lists in Programming?**

Lists are important in programming because they allow us to store and manage multiple values together in one place. Without lists, we would need to use separate variables for each value, making the code messy and harder to maintain.

---
**Use Case:**  
Imagine you need to store 100 student names or 100 exam scores.  
Without lists, you would need to create 100 different variables. With a list, you can store all of them in a single place.

## **List Comprehension** 

List comprehension is a concise way to create lists in Python by applying an expression to each item in an existing iterable (like a list, string, or range).

```python
new_list = [expression for item in iterable if condition]
```


- expression: The operation or value you want to add to the new list.

- item: Each element in the iterable.

- iterable: The collection (e.g., list, range) to loop through.

- condition (optional): A filter to include only items that meet a specific condition.

**Advantages of List Comprehension**

---
- **Shorter and Cleaner Code**:  
  List comprehensions allow you to write more concise code in a single line.

- **Improved Readability**:  
  The syntax is clear and straightforward, making the code easier to understand at a glance.

- **Faster Execution**:  
  List comprehensions are generally faster than using traditional loops for creating lists, as they are optimized for this purpose.

- **Less Memory Usage**:  
  They are more memory-efficient, as Python creates the list in a single step.

- **Flexibility**:  
  You can easily add conditions to filter items, making it more versatile than regular loops.

- **Cleaner Syntax for Nested Loops**:  
  List comprehensions can simplify the process of flattening or handling nested lists.

````python
## Add 1 to 10 Numbers to a list

## Normal way
L=[]

for i in range (1,11):
    L.append(i)

L
````

````python
## Better Way --> One Line

L = [i for i in range(1,11)]

L
````

````python
## Scalar Multiplication on a Vector 

v=[2,3,4]
s=-2

# [-4,-6,-8]

x=[]
for i in v:
    x.append(i*s)
print(x)
````

````python
v=[2,3,4]
s=-2

x=[i*s for i in v]
x
````

````python
## Add Squares

L= [1,2,3,4,5]

[i**2 for i in L]
````

````python
## Print all numbers the divisible by 5 in the range (1 to 50)

x=[i for i in range(1,51) if i%5==0]

print(x)
````

````python
## Find languages which start with letter p 

languages=["java","python","php","c++"]

[i for i in languages if i.startswith('p')]
````

````python
##Nested if in list comprehension

basket=["apple","banana","orange"]
my_fruits=["apple","kiwi","banana"]

## add new list from my_fruits and items if the fruits exists in the basket and starts with "a"

[fruit for fruit in my_fruits]

[fruit for fruit in my_fruits if fruit in basket ]

[fruit for fruit in my_fruits if fruit in basket if fruit.startswith('a')]
````

**Disadvantages of Lists**  
- **Slower for large data**: Lists can become inefficient for large datasets.  
- **High memory usage**: Lists consume more memory compared to other data structures.  
- **Slower search time**: Searching for an item in a large list can be slow.  
- **Limited functionality**: Lists are less powerful than other specialized structures like sets or dictionaries.  
- **Mutable**: Lists can be modified, which can lead to unintentional changes.  
- **Inconsistent data types**: Lists allow mixed data types, making consistency harder to enforce.

### **Apply a Function to a Single Column (Series)**

```python
def rating(x, high, low):
    if x > high:
        return "High"
    elif x > low:
        return "med"        
    else:
        return "low"
```

```python
sales_data['sales_amount'].apply(rating, args = (1800, 1000))  # Apply the function with additional arguments
```
